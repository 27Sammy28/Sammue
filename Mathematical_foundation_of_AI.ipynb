```python
import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
import numpy as np
np.random.seed(3)
rng = np.random.default_rng()
data1 = rng.poisson(5, 100)
data2 = rng.poisson(9, 100)


import matplotlib.pyplot as plt
plt.hist(data1, density=True, ec='red', color='green')
plt.show()
plt.hist(data2, density=True, ec='yellow', color='red')
plt.show()
```

```python
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(data1, data2)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

```python
import numpy as np
np.random.seed(1)
rng = np.random.default_rng()

rng = np.random.default_rng()
n, p = 10, .6  # Just some parameters for binomial distribution
beforeTreatment = rng.binomial(n, p, 40)

afterTreatment = np.array([0]*40)
for i in range(len(beforeTreatment)):
  afterTreatment[i] = beforeTreatment[i] + rng.binomial(5, 0.5, 1) - 4
for i in range(len(beforeTreatment)):
    print(beforeTreatment[i], afterTreatment[i])
```

```python
import matplotlib.pyplot as plt
plt.hist(beforeTreatment, density=True)
plt.show()
plt.hist(afterTreatment, density=True)
plt.show()


plt.scatter(afterTreatment, beforeTreatment)
plt.show()
```

```python
from scipy.stats import wilcoxon

stat, p = wilcoxon(beforeTreatment, afterTreatment)
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
	print('Same distribution (fail to reject H0)')
else:
	print('Different distribution (reject H0)')
```

# The sample size can have an impact on the coefficient of determination. In general, as the sample size increases, the coefficient of determination tends to become more reliable and accurate. This is because a larger sample size allows for a better estimation of the true relationship between X and Y and reduces the influence of random variation. Therefore, increasing the sample size typically results in a higher determination coefficient.
